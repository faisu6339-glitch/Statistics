```python
# âœ… Example 1: Basic PDF & CDF Calculation
```

```python
from scipy.stats import expon
```

```python
# Average rate (Î»)
lmbda = 0.2  # events per minute â†’ mean waiting time = 1/Î» = 5 minutes

# Given time (t)
t = 10  # minutes
```

```python
# PDF (Probability Density Function)
pdf = expon.pdf(t, scale=1/lmbda)
```

```python
# CDF (Cumulative Distribution Function)
cdf = expon.cdf(t, scale=1/lmbda)
```

```python
print(f"Î» = {lmbda}")
print(f"Mean waiting time = {1/lmbda} minutes")
print(f"PDF f({t}) = {pdf:.5f}")
print(f"CDF F({t}) = {cdf:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
# Rate Î» and scale
lmbda = 0.2
scale = 1 / lmbda  # mean
```

```python
# Values for t
t = np.linspace(0, 30, 200)
```

```python
# PDF and CDF
pdf = expon.pdf(t, scale=scale)
cdf = expon.cdf(t, scale=scale)
```

```python
# Plot
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.plot(t, pdf, color='blue')
plt.title(f'Exponential PDF (Î» = {lmbda})')
plt.xlabel('Time (t)')
plt.ylabel('f(t)')
plt.grid(True)
```

```python
plt.subplot(1, 2, 2)
plt.plot(t, cdf, color='green')
plt.title(f'Exponential CDF (Î» = {lmbda})')
plt.xlabel('Time (t)')
plt.ylabel('F(t)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Î» = 0.5 â†’ average of 2 minutes between events
lmbda = 0.5
scale = 1 / lmbda

# Generate 1000 random samples
data = np.random.exponential(scale, 1000)

plt.hist(data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f'Simulated Exponential Distribution (Î» = {lmbda})')
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

def exponential_analysis():
    print("\n=== Exponential Distribution Analysis ===")
    lmbda = float(input("Enter rate (Î»): "))
    scale = 1 / lmbda

    while True:
        print("\n1. PDF (f(t)) at a given time")
        print("2. CDF (F(t)) up to time t")
        print("3. Probability (T > t)")
        print("4. Mean and Variance")
        print("5. Plot PDF and CDF")
        print("6. Simulate random waiting times")
        print("7. Exit")
        choice = input("Select an option: ")

        if choice == '1':
            t = float(input("Enter time (t): "))
            prob = expon.pdf(t, scale=scale)
            print(f"f({t}) = {prob:.5f}")

        elif choice == '2':
            t = float(input("Enter time (t): "))
            prob = expon.cdf(t, scale=scale)
            print(f"F({t}) = {prob:.5f} â†’ P(T â‰¤ {t})")

        elif choice == '3':
            t = float(input("Enter time (t): "))
            prob = 1 - expon.cdf(t, scale=scale)
            print(f"P(T > {t}) = {prob:.5f}")

        elif choice == '4':
            mean = expon.mean(scale=scale)
            var = expon.var(scale=scale)
            print(f"Mean = {mean:.2f}, Variance = {var:.2f}")

        elif choice == '5':
            t = np.linspace(0, 10/ lmbda, 200)
            pdf = expon.pdf(t, scale=scale)
            cdf = expon.cdf(t, scale=scale)

            plt.figure(figsize=(10, 5))
            plt.subplot(1, 2, 1)
            plt.plot(t, pdf, color='blue')
            plt.title('Exponential PDF')
            plt.xlabel('Time (t)')
            plt.ylabel('f(t)')
            plt.grid(True)

            plt.subplot(1, 2, 2)
            plt.plot(t, cdf, color='green')
            plt.title('Exponential CDF')
            plt.xlabel('Time (t)')
            plt.ylabel('F(t)')
            plt.grid(True)

            plt.tight_layout()
            plt.show()

        elif choice == '6':
            n = int(input("Enter number of samples: "))
            data = np.random.exponential(scale, n)
            plt.hist(data, bins=30, color='orange', edgecolor='black', alpha=0.7)
            plt.title(f'Simulated Exponential Data (Î» = {lmbda})')
            plt.xlabel('Time')
            plt.ylabel('Frequency')
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.show()

        elif choice == '7':
            print("Exiting program. Thank you!")
            break
        else:
            print("Invalid option. Try again.")

if __name__ == "__main__":
    exponential_analysis()
```
