```python
#  1. Range
```

```python
range_weight = df["weight"].max() - df["weight"].min()
print("Range of weight:", range_weight)

range_mpg = df["mpg"].max() - df["mpg"].min()
print("Range of mpg:", range_mpg)

```

```python
# 2. Variance
```

```python
print("Variance of weight:", df["weight"].var())
print("Variance of mpg:", df["mpg"].var())
```

```python
# 3. Standard Deviation (SD)
```

```python
print("Standard Deviation of weight:", df["weight"].std())
print("Standard Deviation of mpg:", df["mpg"].std())
```

```python
# 4. Interquartile Range (IQR)
```

```python
Q1_weight = df["weight"].quantile(0.25)
Q3_weight = df["weight"].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight
print("IQR of weight:", IQR_weight)

Q1_mpg = df["mpg"].quantile(0.25)
Q3_mpg = df["mpg"].quantile(0.75)
IQR_mpg = Q3_mpg - Q1_mpg
print("IQR of mpg:", IQR_mpg)
```

```python
# 5. Coefficient of Variation (CV)
```

```python
cv_weight = df["weight"].std() / df["weight"].mean()
cv_mpg = df["mpg"].std() / df["mpg"].mean()

print("Coefficient of Variation (Weight):", cv_weight)
print("Coefficient of Variation (MPG):", cv_mpg)
```

```python


#Visualization of Measures of Dispersion
```

```python
#1. Histogram (Spread of Data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df["mpg"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df["weight"], bins=20, kde=True, color="salmon")
plt.title("Distribution of Weight")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()
```

```python
#2. Boxplot (IQR & Outliers)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df["mpg"], color="lightblue")
plt.title("Boxplot of MPG (Dispersion & Outliers)")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x=df["weight"], color="lightgreen")
plt.title("Boxplot of Weight (Dispersion & Outliers)")
plt.show()
```

```python
# 3. Violin Plot (Spread + Density)
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x=df["mpg"], color="orange")
plt.title("Violin Plot of MPG")
plt.show()

plt.figure(figsize=(10,5))
sns.violinplot(x=df["weight"], color="purple")
plt.title("Violin Plot of Weight")
plt.show()
```

```python
#4. Side-by-Side Boxplot (Comparison)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[["mpg","weight"]])
plt.title("Comparison of Dispersion: MPG vs Weight")
plt.show()
```
