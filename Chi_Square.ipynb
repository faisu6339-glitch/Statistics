```python
import scipy.stats as stats

# Observed frequencies
observed = [54, 98, 63, 7, 5, 4, 5, 88]

# Expected frequencies (optional â€” if all categories expected equally likely)
expected = [sum(observed)/len(observed)] * len(observed)

# Perform Chi-square test
chi_square, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-square test results:")
print(f"Chi-square statistic: {chi_square:.4f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis interpretation
alpha = 0.05
null_hyp = "H0: Observed frequencies match expected frequencies"
alt_hyp = "H1: Observed frequencies do not match expected frequencies"

print(f"Null Hypothesis: {null_hyp}")
print(f"Alternative Hypothesis: {alt_hyp}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis (H0)")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0)")
```

```python
import scipy.stats as stats
import numpy as np

def chi_square_independence_test(contingency_table, alpha=0.05):
    """
    Performs a Chi-square test of independence on a contingency table.

    Parameters:
    contingency_table : 2D list or array : observed counts in a contingency table
    alpha             : float             : significance level (default 0.05)

    Returns:
    dict with keys:
        chi2        : float : Chi-square statistic
        p_value     : float : P-value
        dof         : int   : degrees of freedom
        expected    : array : expected frequencies
        null_hyp   : str   : Null hypothesis
        alt_hyp    : str   : Alternative hypothesis
        conclusion : str   : Test conclusion
    """
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

    null_hyp = "H0: The two categorical variables are independent"
    alt_hyp = "H1: The two categorical variables are dependent"

    if p_value < alpha:
        conclusion = "Reject the null hypothesis (H0)"
    else:
        conclusion = "Fail to reject the null hypothesis (H0)"

    return {
        'chi2': chi2,
        'p_value': p_value,
        'dof': dof,
        'expected': expected,
        'null_hypothesis': null_hyp,
        'alternative_hypothesis': alt_hyp,
        'conclusion': conclusion
    }

# Example usage
# Contingency table: rows = Gender (Male/Female), columns = Pass/Fail
observed_table = [
    [30, 10],  # Male: 30 pass, 10 fail
    [20, 20]   # Female: 20 pass, 20 fail
]

result = chi_square_independence_test(observed_table, alpha=0.05)

print("Chi-square Test of Independence Results:")
print(f"Chi-square statistic: {result['chi2']:.4f}")
print(f"P-value: {result['p_value']:.4f}")
print(f"Degrees of freedom: {result['dof']}")
print("Expected frequencies:")
print(np.round(result['expected'], 2))
print(f"Null Hypothesis: {result['null_hypothesis']}")
print(f"Alternative Hypothesis: {result['alternative_hypothesis']}")
print(f"Conclusion at alpha=0.05: {result['conclusion']}")
```
