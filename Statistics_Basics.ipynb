```python
#Clean categorical columns (strip whitespace, unify case)
```

```python
data["Gender"]   = data["Gender"].astype(str).str.strip().str.upper()
data["Ranking"]  = data["Ranking"].astype(str).str.strip().str.title()
```

```python
gender_counts = data["Gender"].value_counts(dropna=False)
ranking_counts = data["Ranking"].value_counts(dropna=False)
```

```python
print("Gender counts:\n", gender_counts)
print("\nRanking counts:\n", ranking_counts)
```

```python
#Frequency as percentages
```

```python
gender_pct = data["Gender"].value_counts(normalize=True, dropna=False) * 100
ranking_pct = data["Ranking"].value_counts(normalize=True, dropna=False) * 100

print("Gender %:\n", gender_pct.round(2))
print("\nRanking %:\n", ranking_pct.round(2))
```

```python
#Combined frequency table with both count & percent
```

```python
def freq_table(col):
    counts = data[col].value_counts(dropna=False)
    pct = data[col].value_counts(normalize=True, dropna=False) * 100
    return pd.DataFrame({"count": counts, "percent": pct.round(2)})

print("Gender frequency table:\n", freq_table("Gender"))
print("\nRanking frequency table:\n", freq_table("Ranking"))
```

```python
#Frequency distribution for marks ranges (bins)
```

```python
bins = [0, 10, 20, 30]   # change ranges if you want
labels = ["0-10", "11-20", "21-30"]

for subj in ["Data Structure", "Mathematics", "Operating System"]:
    binned = pd.cut(data[subj], bins=bins, labels=labels, include_lowest=True)
    freq = binned.value_counts().sort_index()
    pct = binned.value_counts(normalize=True).sort_index() * 100
    print(f"\n{subj}  counts by bin:\n{pd.concat([freq, pct.round(2)], axis=1).rename(columns={0:'count',1:'percent'})}")
```

```python
#Cross-tabulation (e.g., Gender vs Ranking)
```

```python
ct = pd.crosstab(data["Gender"], data["Ranking"], margins=True)   # margins=True adds totals
print("Gender x Ranking counts:\n", ct)

ct_pct = pd.crosstab(data["Gender"], data["Ranking"], normalize="index") * 100
print("\nGender x Ranking percent (row-wise):\n", ct_pct.round(2))
```

```python


# Bar chart for Gender counts
gender_counts.plot(kind="bar", rot=0)
plt.title("Gender Frequency")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Pie for Ranking
ranking_counts.plot(kind="pie", autopct="%1.1f%%", ylabel="")
plt.title("Ranking Distribution")
plt.show()

# Stacked bar: Ranking by Gender
pd.crosstab(data["Gender"], data["Ranking"]).plot(kind="bar", stacked=True)
plt.title("Ranking by Gender (stacked)")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Ranking")
plt.show()
```
