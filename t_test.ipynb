#Independent t-Test

```python
from scipy.stats import ttest_ind
import numpy as np

# Feature values for two classes
class_0 = np.array([10, 12, 11, 9, 10])
class_1 = np.array([18, 20, 17, 19, 21])

t_stat, p_value = ttest_ind(class_0, class_1)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

#Paired t-Test (Model Comparison)

```python
from scipy.stats import ttest_rel

before = [78, 80, 79, 77, 76]
after  = [82, 85, 83, 84, 86]

t_stat, p_value = ttest_rel(before, after)

print("p-value:", p_value)
```

```python
from sklearn.datasets import load_iris
from scipy.stats import ttest_ind

X, y = load_iris(return_X_y=True)

# Compare feature 0 between class 0 and class 1
feature = X[:, 0]

class0 = feature[y == 0]
class1 = feature[y == 1]

t_stat, p = ttest_ind(class0, class1)
print("p-value:", p)
```
